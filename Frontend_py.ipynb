```python
import React, { useState } from 'react';
import Select from 'react-select';
import axios from 'axios';

const App = () => {
  const [jsonInput, setJsonInput] = useState('');
  const [error, setError] = useState('');
  const [responseData, setResponseData] = useState({});
  const [selectedOptions, setSelectedOptions] = useState([]);

  const options = [
    { value: 'alphabets', label: 'Alphabets' },
    { value: 'numbers', label: 'Numbers' },
    { value: 'highestLowercase', label: 'Highest lowercase alphabet' }
  ];

  const handleSubmit = async () => {
    try {
      const parsedData = JSON.parse(jsonInput);
      setError(''); // Clear previous errors

      const response = await axios.post('YOUR_API_ENDPOINT', parsedData);
      setResponseData(response.data);
    } catch (err) {
      setError('Invalid JSON input');
    }
  };

  const handleDropdownChange = (selected) => {
    setSelectedOptions(selected);
  };

  const renderResponse = () => {
    if (!responseData) return null;

    const displayData = {};
    selectedOptions.forEach((option) => {
      if (option.value === 'alphabets') displayData.alphabets = responseData.alphabets;
      if (option.value === 'numbers') displayData.numbers = responseData.numbers;
      if (option.value === 'highestLowercase') displayData.highestLowercase = responseData.highestLowercase;
    });

    return <pre>{JSON.stringify(displayData, null, 2)}</pre>;
  };

  return (
    <div>
      <h1>Your Roll Number as Title</h1>
      <textarea
        value={jsonInput}
        onChange={(e) => setJsonInput(e.target.value)}
        placeholder='Enter JSON here'
      />
      <button onClick={handleSubmit}>Submit</button>
      {error && <p style={{ color: 'red' }}>{error}</p>}

      {responseData && (
        <>
          <Select
            isMulti
            options={options}
            onChange={handleDropdownChange}
          />
          {renderResponse()}
        </>
      )}
    </div>
  );
};

export default App;
```
